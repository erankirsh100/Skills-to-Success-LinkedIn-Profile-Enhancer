```python
import google.generativeai as genai
import os

# INSERT A LIST OF API KEYS TO GENAI HERE
API_LIST = []


with open('running_index_skills.txt', 'r') as f:
    running_index = int(f.read())

if 'job_skills.pkl' not in os.listdir():
    job_skills = pd.DataFrame(columns = ['company_name', 'job', 'skills'])
else:
    job_skills = pd.read_pickle('job_skills.pkl')

for company_row in scraped_compnies.iloc[running_index:].iterrows():
    cur_company_name = company_row[1]['company_name']
    cur_job = company_row[1]['title']
    cur_description = company_row[1]['description']

    os.environ["API_KEY"] = API_LIST[api_list_index]
    genai.configure(api_key=os.environ["API_KEY"])

    model = genai.GenerativeModel('gemini-1.5-flash-latest')
    response = model.generate_content(start_prompt(cur_description))
    cur_skills = response.text

    # Create a new DataFrame with the new row
    new_row = pd.DataFrame([{'company_name': cur_company_name, 'job': cur_job, 'skills': cur_skills}])

    # Concatenate the new row to the existing DataFrame
    job_skills = pd.concat([job_skills, new_row], ignore_index=True)

    # Save the updated DataFrame back to the pickle file
    job_skills.to_pickle('job_skills.pkl')

    running_index += 1
    api_list_index += 1

    if api_list_index == len(API_LIST):
        api_list_index = 0

    with open('running_index_skills.txt', 'w') as f:
        f.write(str(running_index))

    # time.sleep(1)

    print(f"Finished {running_index}, {cur_company_name}, {cur_job}")

# os.environ["API_KEY"] = 'AIzaSyDV6Ksr-PfwQobqys_ed8K9HxzFNRrlhlU'
# genai.configure(api_key=os.environ["API_KEY"])

# model = genai.GenerativeModel('gemini-1.5-flash-latest')
# response = model.generate_content(start_prompt(scraped_compnies['description'][0]))
# print(response.text)
```

```python
import re

def parse_job_description(job_description):
    # job_description contains many <p> tags, and list tags, we want to remove them and get the text
    job_description = job_description.replace('<p>', '').replace('</p>', '').replace('<li>', '').replace('</li>', '')
    job_description = job_description.replace('<ul>', '').replace('</ul>', '').replace('<ol>', '').replace('</ol>', '')
    job_description = job_description.replace('<div>', '').replace('</div>', '')
    job_description = job_description.replace('<br>', ' ')
    # stripping the leading, trailing, and extra whitespaces
    job_description = job_description.strip()
    job_description = job_description.replace('<div id="jobDescriptionText" class="jobsearch-jobDescriptionText">', '')
    job_description = job_description.replace('</b>', '').replace('<b>', '')
    job_description = job_description.replace('<i>', '').replace('</i>', '')
    # removing more than 2 newlines in a row
    job_description = re.sub('(\n\n)\n*|\n', r'\1', job_description)
    return job_description


# indeed_indeed_kaggle_1 is the raw data from kaggle
indeed_kaggle = pd.read_csv('indeed_kaggle_1.csv')

indeed_kaggle = indeed_kaggle[['Company Name', 'Job Title', 'Apply Url', 'Location', 'Salary From', 'Job Type', 'Job Description']]

# parsing the job description
indeed_kaggle['Job Description'] = indeed_kaggle['Job Description'].apply(parse_job_description)

display(indeed_kaggle[['Company Name', 'Job Title', 'Apply Url', 'Location', 'Salary From', 'Job Type', 'Job Description']])

# save to parquet
indeed_kaggle.to_parquet('indeed_kaggle.parquet')
```
